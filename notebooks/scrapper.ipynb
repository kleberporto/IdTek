```python
import requests
from bs4 import BeautifulSoup

URL = "https://poatek.com/our-team"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find_all("div", {"class": "team team_vertical"})
print(results)
```

```python
print(results[0])
```
